| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Subset the DataFrame for type "Movie"
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]

# Start by filtering out movies that were released before 1990
movies_90s = netflix_subset[(netflix_subset["release_year"] >= 1990) & (netflix_subset["release_year"] <= 2000)]
print(movies_90s)
```

```python
# Visualize the duration column of filtered data of the movies with more of 90 minutes

plt.hist(movies_90s["duration"])
plt.title("Movies Duration in the 90s")
plt.xlabel("Duration in minutes")
plt.ylabel("Numbers of Movies")

plt.show
```

```python
# Now we will continue filtering only action movies
action_movies_90s = movies_90s[movies_90s["genre"] == "Action"]

# Ensure the 'duration' column is of integer type
action_movies_90s['duration'] = action_movies_90s['duration'].astype(int)

# We want to know the amount of the movies shorter than 90 minutes
short_action_movies = (action_movies_90s['duration'] < 90).sum()
short_action_movies
```
